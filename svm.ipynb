```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from tensorflow.keras.datasets import cifar10
```

```python

# Load CIFAR-10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python

# Flatten images to 1D vectors
x_train_flat = x_train.reshape(x_train.shape[0], -1) / 255.0
x_test_flat = x_test.reshape(x_test.shape[0], -1) / 255.0
```

```python

# Use only a subset for faster training (SVMs are slow on large data)
x_train_small = x_train_flat[:5000]
y_train_small = y_train[:5000].ravel()
x_test_small = x_test_flat[:1000]
y_test_small = y_test[:1000].ravel()
```

```python

# Create and train an SVM classifier
svm_model = SVC(kernel='poly')  # You can try 'rbf' or 'poly' as well
svm_model.fit(x_train_small, y_train_small)
```

```python

# Predict and evaluate
y_pred = svm_model.predict(x_test_small)
acc = accuracy_score(y_test_small, y_pred)
print(f"SVM Accuracy on CIFAR-10 (subset): {acc:.4f}")
```

```python
import joblib
joblib.dump(svm_model, 'svm_model1.pkl')
```
